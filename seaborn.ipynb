# Seaborn 

dataset used = iris 

```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns 
```

```python
#sns.set()

#switches to seaborn default 
```

```python
df = pd.read_csv('iris.csv', index_col=0)

#index_col = 0, we are explicitly telling to treat the 1st column as index 
#index_col = False exact opposite 
```

```python
#default is scatter plot, no dedicated scatter ()
#lmplot = line plot 

sns.lmplot(x='sepal_length', y='sepal_width', data=df, fit_reg=False, hue='species')
#hue = determines which category the points belong to aka target variable 

sns.set_style('whitegrid')
```

```python
sepal_df = df.drop(['petal_width', 'petal_length'], axis=1)
petal_df = df.drop(['sepal_length', 'sepal_width'], axis=1)

sns.boxplot(data=sepal_df)
sns.set_style('whitegrid')
```

```python
#violin plot = alterbative to box plot 

sns.violinplot(x='species', y='sepal_length', data = df, palette=['red', 'blue', 'pink'])
plt.ylim(3.8, None)

#in sns inner=None = remove inner bars 
```

```python
#simply display each point 

sns.swarmplot(x='species', y='petal_length', data = df, palette=['red', 'blue', 'pink'], split=True)
sns.set_style('whitegrid')
```

```python
sns.swarmplot(x='petal_width', y='petal_length', data = df, palette=['red', 'blue', 'pink'], hue='species', split=True)
```

```python
sns.violinplot(x='species', y='petal_length', data = df, inner = None)

sns.swarmplot(x='species', y='petal_length', data = df, color='black', alpha=0.7)
```

Heatmap

```python
corr = petal_df.corr()
sns.heatmap(corr)

#plt.figure(figsize=(5,10))
```

```python
#distribution plot aka histogram 

sns.distplot(df.petal_length)
sns.distplot(df.petal_width)
```

```python
#count plot aka bar plot 

sns.countplot(y='sepal_width', data=df)
```
