# Pie chart

```python
import matplotlib.pyplot as plt 
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
label = 'frogs', 'dogs', 'logs', 'mops' #anticlockwise 
portion = [30, 40, 10, 20]
explode = (0, 0.08, 0, 0)
colors= ['yellow', 'red', 'blue', 'green']

#plt.pie(size, labels, autopct)
plt.pie(portion, autopct='%1.0f%%', shadow=True, explode=explode, startangle=0, colors=colors)

plt.legend(label,loc='best', shadow=True)
plt.axis('equal') #ensure pie is drawn as a well fit circle ('scaled') ('off')
plt.title('my very own pie chart', fontsize=14)
plt.tight_layout() #fix pie chart into a better figure 
```

```python
zoo = pd.read_csv('zoo.csv')

water_need = zoo['water_need'].head(5).reset_index(drop=True)
animal = zoo['animal'].head(5).reset_index(drop=True)
```

```python
plt.pie(water_need, labels=animal, autopct='%1.1f%%', startangle=0)
plt.title('zoo')
```
