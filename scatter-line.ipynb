```python
import numpy as np
```

```python
import matplotlib.pyplot as plt 
```

axes - plotting area 
axis - limit, set points on x,y 
artist - every single component everything : axes, grid, limits, legends 

```python
fig, ax = plt.subplots()
#plt.subplot is a function which returns tuples of figure and axes(not axis). fig and ax unpack these tuples into variables 
#default 1.0 to 4.0 

ax.plot([1,2,3,4], [2,1,4,3])
#execute in same cell 
```

```python
plt.plot([1,2,3,4], [4,2,3,1], 'b^')
#no need to create an explicit ax axes as before 
# r-- : red dash 
# ro : red spots 
# r^ : red triangle 
# rs : red square 
# rx- red x 
# r,b,g colors 


plt.axis([0,5,0,5])
#xmin, xmax, ymin, ymax 
```

```python
y = np.linspace(1,10)
#linspace will create a 100 points from 1 to 10 

y = np.linspace(1,10,10)
#it will create 10 points from 1 to 10 
```

```python
%matplotlib inline

x= np.linspace(0,2,10)

plt.plot(x,x,'bs', label='linear')
plt.plot(x,x**2, 'go', label='sqaured')
plt.plot(x,x**3, 'r--', label='cubic')
#the color and shape have to come after x,x plotting only 

plt.xlabel('x-label')
plt.ylabel('y-label')

plt.legend()

plt.title('the x factor')
```

```python
plt.plot([1,2,3,4])
#when given a single list of values, matplot asuumes it to be y values i.e 1 to 4 = 4 values 
#and auto matically generates x values, as python default starts from 0 hence here x is from 0 to 3 = 4 values as well 

plt.show()
#displays figure
#but here no use as we are in notebook 
```

```python
t = np.arange(0, 12, 2)
#(start, stop, step)
#it wont count 12 

plt.plot(t,t, 'b^')
plt.plot(t, t**2, 'ro')
plt.plot(t, t**3, 'gs')
#here the y keeps expanding to get its equivalent of x 
```

```python
x= ['A', 'B', 'C', 'D', 'E']
y= [20,40,50,10,30]

plt.figure(figsize=(11,3))
#(x-length, y-length)

plt.subplot(131)
#1x3 grid   1st subplot 
plt.bar(x,y)

plt.subplot(132)
plt.plot(x,y)

plt.subplot(133)
plt.scatter(x,y)

plt.suptitle('the x factor')
```
